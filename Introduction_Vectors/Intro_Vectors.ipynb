#### Manual Solution No. 4

a. $\textbf{u} = (5, -7)$
$$
\begin{align}
    \textbf{u}' = \frac{\textbf{u}}{||\textbf{u}||} &= \frac{(5, -7)}{\sqrt{5^2 + (-7)^2}} \\
        &= \frac{(5, -7)}{\sqrt{25 + 49}} \\

\end{align}
$$


b. $\textbf{v} = (1, 2, -2, 4)$
$$
\begin{align}

        &= \frac{(1, 2, -2, 4)}{\sqrt{1 + 4 + 4 + 16}} \\
        &= \frac{(1, 2, -2, 4)}{\sqrt{25}} \\
        &= \frac{(1, 2, -2, 4)}{5} \\
        &= (\frac{1}{5}, \frac{2}{5}, -\frac{2}{5}, \frac{4}{5}) 
\end{align}
